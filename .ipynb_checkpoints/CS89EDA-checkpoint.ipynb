## Exploratory Data Visualizations

These are histogram plots of all features


## Histogram Plots of All Features

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import pandas as pd
pd.__version__

df = pd.read_csv("dataset.csv", sep=',', encoding='latin-1')
df.info()
d={"negative":0, "positive":1,"not_detected":0,"detected":1}
df=df.replace(d)
df_num=df.select_dtypes(include=['float64', 'int64'])
df_num.hist(figsize=(16,20),bins=50,xlabelsize=5,ylabelsize=5)
```

## Identifying Strongly Correlated Features

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import pandas as pd
pd.__version__

df = pd.read_csv("dataset.csv", sep=',', encoding='latin-1')

df.info()
d={"negative":0, "positive":1,"not_detected":0,"detected":1}
df=df.replace(d)
df.info()
df_num=df.select_dtypes(include=['float64', 'int64'])
df_num_corr = df_num.corr()['SARS-Cov-2 exam result'][:-1]
feature_list=df_num_corr[abs(df_num_corr)>.2].sort_values(ascending=False)
print("there are {} correlated values with sars-cov-2 exam result:\n{}".format(len(feature_list),feature_list))
#df_num_corr.hist(figsize=(16,20),bins=50,xlabelsize=5,ylabelsize=5)
f_list=df_num_corr[1]
print(f_list)
corr=df_num.drop('SARS-Cov-2 exam result', axis=1).corr()
```
